```python
%config IPCompleter.greedy=True

import glob
import pandas as pd
import numpy as np
import re
import os
```

```python
results_dir = 'results_m'
results_dir_glob = '/mnt/1717A37A1971CE02/WorkSpaces/BDMA/database-system-architecture/project-cmunizcu-damjad-econte/project-cmunizcu-damjad-econte-benchmarks/'+results_dir+'/results*'
final_result = '/mnt/1717A37A1971CE02/WorkSpaces/BDMA/database-system-architecture/project-cmunizcu-damjad-econte/project-cmunizcu-damjad-econte-benchmarks/'+results_dir+'/final.csv'
```

```python
result_files = glob.glob(results_dir_glob)
```

```python
def read_csv(fileName):
    df = pd.read_csv(fileName)
    m = re.search('results_(?P<n>\d+)_(?P<m>\d+)_(?P<d>\d+)\.csv',fileName)
    if (df.shape[0]==0):
        s2 = pd.Series([np.nan for n in df.columns.values], index=df.columns)
        df = df.append(s2, ignore_index = True)
    df['N'] = int(m.group('n'))
    df['M'] = int(m.group('m'))
    df['d'] = int(m.group('d'))
    return df
```

```python
frames = []
for f in result_files:
    if os.stat(f).st_size != 0:
        df = read_csv(f)
        frames.append(df)
```

```python
df = pd.concat(frames, ignore_index=True)
df = df.sort_values(['N', 'M', 'd'])
#df = df.reset_index()
df.head()
```

```python
df.to_csv(final_result)
```
